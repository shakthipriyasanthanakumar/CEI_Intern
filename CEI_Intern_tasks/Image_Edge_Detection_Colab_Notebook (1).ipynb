```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image
import io
```

```python
# Upload your image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python
# Read and process the image
image = Image.open(image_path).convert('RGB')
image_np = np.array(image)

# Convert to grayscale
gray = cv2.cvtColor(image_np, cv2.COLOR_RGB2GRAY)

# Apply Canny Edge Detection
edges = cv2.Canny(gray, 100, 200)
```

```python
# Plot results
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

axs[0].imshow(image_np)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(gray, cmap='gray')
axs[1].set_title('Grayscale Image')
axs[1].axis('off')

axs[2].imshow(edges, cmap='gray')
axs[2].set_title('Canny Edge Detection')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```

```python
# ðŸ“Œ Install OpenCV if not already
!pip install opencv-python-headless

# ðŸ“Œ Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image
import io

# ðŸ“Œ Upload an image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# ðŸ“Œ Load and prepare image
image = Image.open(image_path).convert('RGB')
image_np = np.array(image)

# Step 1: Convert to grayscale
gray = cv2.cvtColor(image_np, cv2.COLOR_RGB2GRAY)

# Step 2: Median blur
gray_blur = cv2.medianBlur(gray, 7)

# Step 3: Detect edges
edges = cv2.adaptiveThreshold(
    gray_blur, 255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    blockSize=9, C=2
)

# Step 4: Smooth color using bilateral filter
color = cv2.bilateralFilter(image_np, d=9, sigmaColor=300, sigmaSpace=300)

# Step 5: Combine edges and color
cartoon = cv2.bitwise_and(color, color, mask=edges)

# ðŸ“Œ Show all stages
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
axs[0].imshow(image_np)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(edges, cmap='gray')
axs[1].set_title('Edge Detection')
axs[1].axis('off')

axs[2].imshow(cartoon)
axs[2].set_title('Cartoonified Image')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
